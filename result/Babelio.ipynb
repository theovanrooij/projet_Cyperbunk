```python
df.drop(df.query("Date_Book == '(30/11/-1)'").index,inplace=True)
df.drop(df.query("Date_Book == 'Existeenéditionaudio'").index,inplace=True)
df.drop(df.query("Year == 1900").index,inplace=True)
```

```python
df
```

```python
df["Date_Book"] = df["Date_Book"].apply(lambda x : x.replace("\n","").replace("\t","").replace(" ",""))
```

```python
df["Year"] = df["Date_Book"].apply(lambda x: datetime.datetime.strptime(x,"(%d/%m/%Y)").year) 
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x="Year",bins=20)
```
